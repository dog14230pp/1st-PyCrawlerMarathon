```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
rdataNoPop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data = json.loads(r.text)
dataNoPop = json.loads(rdataNoPop.text)
print(data)
```

```python
print("è³‡æ–™æ•¸é‡:", len(data))
print("æ¬„ä½: ", list(data[0].keys()))
```

```python
#å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for d in data:
    print("æ¨™é¡Œ: ", d['title'])
    print("è²¼æ–‡æ™‚é–“: ", d['createdAt'])
    print("ç•™è¨€äººæ•¸: ", d['commentCount'])
    print("æŒ‰è®šäººæ•¸: ", d['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
dataMeanC = 0 #for popular comment
dataNoPopMeanC = 0 #for Non-popular comment
dataMeanL = 0 #for popular Like
dataNoPopMeanL = 0 #for Non-popular Like

for d in data:
    dataMeanC +=  d['commentCount']
    dataMeanL += d['likeCount']

for d in dataNoPop:
    dataNoPopMeanC +=  d['commentCount']
    dataNoPopMeanL += d['likeCount']
    
print("ç†±é–€å¹³å‡ç•™è¨€äººæ•¸: ", dataMeanC/len(data))
print("ç†±é–€å¹³å‡æŒ‰è®šäººæ•¸: ", dataMeanL/len(data))
print("ç†±é–€å¹³å‡ç•™è¨€äººæ•¸: ", dataNoPopMeanC/len(dataNoPop))
print("ç†±é–€å¹³å‡æŒ‰è®šäººæ•¸: ", dataNoPopMeanL/len(dataNoPop))
```
