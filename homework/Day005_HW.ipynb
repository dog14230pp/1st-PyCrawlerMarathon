## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
import json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
rdataNoPop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data = json.loads(r.text)
dataNoPop = json.loads(rdataNoPop.text)
print(data)
```

```python
print("資料數量:", len(data))
print("欄位: ", list(data[0].keys()))
```

```python
#取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for d in data:
    print("標題: ", d['title'])
    print("貼文時間: ", d['createdAt'])
    print("留言人數: ", d['commentCount'])
    print("按讚人數: ", d['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
dataMeanC = 0 #for popular comment
dataNoPopMeanC = 0 #for Non-popular comment
dataMeanL = 0 #for popular Like
dataNoPopMeanL = 0 #for Non-popular Like

for d in data:
    dataMeanC +=  d['commentCount']
    dataMeanL += d['likeCount']

for d in dataNoPop:
    dataNoPopMeanC +=  d['commentCount']
    dataNoPopMeanL += d['likeCount']
    
print("熱門平均留言人數: ", dataMeanC/len(data))
print("熱門平均按讚人數: ", dataMeanL/len(data))
print("熱門平均留言人數: ", dataNoPopMeanC/len(dataNoPop))
print("熱門平均按讚人數: ", dataNoPopMeanL/len(dataNoPop))
```
