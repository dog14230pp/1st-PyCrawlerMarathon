### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text)
for tmp in soup.find(class_ = 'part_list_2').find_all('h3'):
    print(tmp.find(class_ = 'date').text, tmp.find('em').text, tmp.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
#We can use the structure dict to store the category
category = {}
for tmp in soup.find(class_ = 'part_list_2').find_all('h3'):
    t = tmp.find('em').text
    if t not in category.keys():
        category[t] = 0
    category[t] += 1
category
```

### ③ 哪一個類別的文章最多

```python
result = max(category, key=category.get)
print("文章之類別最多者為:", result, "，共", category[result])
```
